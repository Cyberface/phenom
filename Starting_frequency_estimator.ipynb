```python
import lal
import lalsimulation as ls
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = []
q = 8.0
Mtotal = 100.0
theta = 60.0
chi = 0.8
m1_SI = q * Mtotal * lal.MSUN_SI / (1.0+q)
m2_SI = Mtotal * lal.MSUN_SI / (1.0+q)
chi1z = chi * np.cos(theta * np.pi / 180.0)
chi2z = 0.0
w = [0.019,0.0195,0.02,0.0198,0.02,0.021,0.03]
for i in w:
    f = i / np.pi / (Mtotal * lal.MTSUN_SI)
    time = ls.SimIMRPhenomDChirpTime(m1_SI,m2_SI,chi1z,chi2z,f) / (Mtotal*lal.MTSUN_SI)
    t.append(time)
```
